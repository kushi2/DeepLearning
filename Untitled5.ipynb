```python
b =[]
f = []
for i in range(0 , title.size):
    #print(i)
    a = str(title[i]).replace("'","")
    b = a.replace("[","")
    b = b.replace("]","")

    f.append(b)
    
print(f)
```

```python
ff = np.array(f)
ff
```

```python
# Importing all necessary libraries 
import cv2 
import os 
final = []
for i  in range(0,len(title)):
#for i in range(0,2):
    print(i)
    
    s = str(ff[i]).replace("'","")
    s = s.replace("[","")
    s = s.replace("]","")
    print(s)
    print(type(s))
    dd = s+".mp4"
    print(dd)
# Read the video from specified path 
    cam = cv2.VideoCapture(dd) 
  
    try: 
      
    # creating a folder named data 
        if not os.path.exists('data'): 
            os.makedirs('data') 
        
#not created then raise error 
    except OSError: 
        print ('Error: Creating directory of data') 
    
#frame 
    currentframe = 0
    imag =[]
    while(True): 
      
    # reading from frame 
        ret,frame = cam.read() 
  
        if ret: 
        # if video is still left continue creating images 
            name =  './data/frame'+ str(video_id[i]) + str(currentframe) + '.jpg'
         
        # increasing counter so that it will 
        # show how many frames are created 
            currentframe += 1
        
        #CAPTURE FRAME, when conditon is true 
            capture_frame = 500
            if currentframe % capture_frame == 0:
                print ('Creating...' + name) 
            # writing the extracted images 
                cv2.imwrite(name, frame) 
                imag.append(name)
             
           
        else: 
            break
            
    final.append(imag)
#Release all space and windows once done 
    cam.release() 
    cv2.destroyAllWindows() 
    
```

```python
fin = np.array(final)
fin
fin = fin.reshape(-1,1)
new_new = np.concatenate((data,fin), 1)
```

```python
x = np.array(image2)
x = x.reshape(len(image2),28,28,3,1)
y = cate1
print(x.shape)
print(y.shape)
y = y.astype(int)
```

```python
#train_test_split will randomly split the data 
xtrain , xtest , ytrain , ytest  =train_test_split(x,y,test_size=0.2) 
print(len(xtrain))
print(len(ytrain)) 
print(len(xtest))
print(len(ytest))
```

```python
xtrain_new = xtrain/255
xtest_new = xtest/255
```

```python
one_hot_y_train = tensorflow.keras.utils.to_categorical(ytrain, num_classes=cat+1)
one_hot_y_test = tensorflow.keras.utils.to_categorical(ytest, num_classes= cat+1)
```

```python
model2 = Sequential()
model2.add(Conv3D(40,kernel_size=(5,5, 5), strides=(2,2,2),
                 activation='relu',padding ='same',
                input_shape = (28,28,3,1)))

model2.add(MaxPooling3D(pool_size=(3, 3 ,1)))

model2.add(Conv3D(40,kernel_size=(5,5, 5), strides=(2,2,2),
                 activation='relu',padding ='same'))
           
model2.add(MaxPooling3D(pool_size=(2, 2 ,1)))





model2.add(Flatten())
model2.add(Dense( cat+1, activation='softmax'))
model2.summary()
```

```python
model2.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history1 = model2.fit(xtrain_new,one_hot_y_train, epochs=50,batch_size=10 ,validation_split=0.2) 
```

```python
import matplotlib.pyplot as plt

plt.plot(history1.history['loss'])

plt.title('Epochs vs Losses ')
plt.ylabel('losses')
plt.xlabel('Epoch')

plt.show()
```
